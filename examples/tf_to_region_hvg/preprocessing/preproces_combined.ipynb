### read example from 10x multiome output

```python
data_path = '../../data/TF_to_region/DEEP_SCENIC/data/'
data_path_processed = '../../data/TF_to_region'
file_name = 'TF_to_region_hvg'
# file_name = 'TF_to_region_marker_genes'
# file_name = 'TF_to_region'
# file_name = 'TF_to_region_ctx'
# file_name = 'TF_to_region_scplus'
# file_name = 'region_accesibility'
seq_len = 768

tsv_file = data_path + file_name + '.tsv'
```

```python
if file_name == 'TF_to_region_scplus':
    tf_to_region = pd.read_csv(tsv_file, delimiter=",", index_col=0)
else:
    tf_to_region = pd.read_csv(tsv_file, delimiter="\t", index_col=0)
# file_name += "_min1"
tf_to_region
```

```python
tf_to_region.fillna(0, inplace=True)
```

```python
if file_name == 'region_accesibility':
    tf_to_region = tf_to_region.T
tf_to_region
```

```python
import re
peak = [re.split("[\-:]+", elem) for elem in tf_to_region.index.to_list()]
peak = pd.DataFrame(peak, columns=['chr', 'start', 'end'])
peak
```

`experimental_id.cell_line.transcription_factor.motif_model`  (motif model is the PWM used to find direct targets of the TF).
example for TF YY1, you have multiple cell lines (e.g. GM12878_female_B-cells_lymphoblastoid_cell_line, lymphoblastoid_cells, …) for each cell line you can also have multiple experimental replicates (e.g. for lymphoblastoid_cells: EXP038605, EXP038618, EXP038614…)

```python
TFs = [re.split("(?<!\d)\.|\.(?!\d)", elem) for elem in tf_to_region.columns.to_list()]
if len(TFs[0]) == 1:
    TFs = pd.DataFrame(TFs, columns=['TF'])
else:
    TFs = pd.DataFrame(TFs, columns=['exp_id', 'cell_line', 'TF', 'motif_model'])
TFs
```

```python
adata = sc.AnnData(tf_to_region.T)
adata.var['chr'] = peak['chr'].values
adata.var['start'] = peak['start'].values
adata.var['end'] = peak['end'].values
# adata = sc.AnnData(tf_to_region)
# adata.obs['chr'] = peak['chr'].values
# adata.obs['start'] = peak['start'].values
# adata.obs['end'] = peak['end'].values
adata
adata.obs['TF'] = TFs['TF'].values
```

```python
def grouped_obs_mean(adata, group_key):
    getX = lambda x: x.X

    grouped = adata.obs.groupby(group_key)
    out = pd.DataFrame(
        np.zeros((adata.shape[1], len(grouped)), dtype=np.float64),
        columns=list(grouped.groups.keys()),
        index=adata.var_names
    )

    for group, idx in grouped.indices.items():
        X = getX(adata[idx])
        out[group] = np.ravel(X.mean(axis=0, dtype=np.float64))

    out = out.where(out > 0, other=0)
    out = out.where(out == 0, other=1)
    return out
```

```python
pbmc_grouped = grouped_obs_mean(adata, "TF")
pbmc_grouped
```

```python
adata_grouped = sc.AnnData(pbmc_grouped.T)
adata_grouped.var['chr'] = peak['chr'].values
adata_grouped.var['start'] = peak['start'].values
adata_grouped.var['end'] = peak['end'].values
adata_grouped.obs['TF'] = pbmc_grouped.columns
```

```python
sc.pp.filter_cells(adata_grouped, min_genes=0)
sc.pp.filter_genes(adata_grouped, min_cells=0)
adata_grouped
```
