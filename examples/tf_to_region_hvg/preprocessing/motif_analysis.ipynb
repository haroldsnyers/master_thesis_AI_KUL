```python
import pandas as pd
import scanpy as sc
from scbasset.scbasset_utils import *

import warnings
warnings.filterwarnings(action='ignore', category=FutureWarning)
```

```python
start_directory = '../../../'
tsv_file = start_directory + 'data/Motif_collection.tsv'

seq_len = 768
type_data = 'TF_to_region'
file_name = 'TF_to_region_ctx'
# file_name = 'TF_to_region_scplus'

data_path = start_directory + 'data/' + type_data + '/processed/'

ad_file = data_path + file_name + '-' + str(seq_len) + '-ad.h5ad'
```

```python
tf_to_region = pd.read_csv(tsv_file, delimiter="\t", index_col=0)
tf_to_region
```

```python
# read h5ad file
ad = anndata.read_h5ad(ad_file)
ad
```

```python
col = ['Motif_name', 'Human_Direct_annot',	'Cluster_Human_Direct_annot']
data = tf_to_region[col]
data = data[data['Motif_name'].apply(lambda x: 'jaspar' in x)]
data
```

```python
TFs = ad.obs.TF.values
list(TFs)
```

```python
df = data[data['Human_Direct_annot'].apply(lambda x: np.any(np.in1d(x, list(TFs))))]
df
```

```python
motif_list = []
for TF in ad.obs.TF:
    TF_index = df.index[df['Human_Direct_annot'] == TF].tolist()
    motif = df['Motif_name'][TF_index[0]].split('__')[1] if len(TF_index) > 0 else 'None'
    motif_list.append(motif)

ad.obs['motif_model'] = motif_list
ad.obs
```

```python
ad.obs.motif_id.value_counts()
```

```python
data_path_processed = start_directory + 'data/TF_to_region'
ad.write('%s/processed/%s-ad.h5ad'%(data_path_processed, file_name))
```
