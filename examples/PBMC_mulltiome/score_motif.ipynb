```python
import anndata
import tensorflow as tf
import numpy as np
import h5py
import matplotlib.pyplot as plt
import os
import math
import pickle
import seaborn as sns
import scipy
import sys
import scanpy as sc
import pandas as pd

from scbasset.utils import *
```

### download data

```python
import subprocess
import os

download_savepath = '../../data/download'
os.makedirs(download_savepath, exist_ok=True)

# download trained scBasset model
if not os.path.exists('%s/pbmc_multiome_best_model.h5'%download_savepath):
    subprocess.run('wget -P %s https://storage.googleapis.com/scbasset_tutorial_data/pbmc_multiome_best_model.h5'%download_savepath, shell=True)

# download motif injection fasta files for CISBP-1.0 motifs and unzip:
if not os.path.exists('%s/Homo_sapiens_motif_fasta.tar.gz'%download_savepath):
    subprocess.run('wget -P %s https://storage.googleapis.com/scbasset_tutorial_data/Homo_sapiens_motif_fasta.tar.gz'%download_savepath, shell=True)
subprocess.run('tar -xzf %s/Homo_sapiens_motif_fasta.tar.gz -C %s/'%(download_savepath, download_savepath), shell=True)
```

### path to input data

```python
trained_model = '../../data/download/pbmc_multiome_best_model.h5'
motif_fasta_folder = '../../data/download/Homo_sapiens_motif_fasta'
rna_data = '../../data/multiome_example/other/ad_rna_annotate.h5ad'
```

### load the data and trained model

```python
# read the cell annotations based on scRNA
ad = anndata.read_h5ad(rna_data)
sc.pp.normalize_total(ad, target_sum=1e6)
sc.pp.log1p(ad)

# load model
model = make_model(32, ad.shape[0], show_summary=False)
model.load_weights(trained_model)
```

### score a TF of interest

```python
scores = motif_score('PAX5', model, motif_fasta_folder=motif_fasta_folder)
ad.obs['PAX5_activity'] = scores
```

### plot cell annotation (left), TF expression (mid) and TF activity (right)

```python
f, axs = plt.subplots(ncols=3, figsize=(15, 4))
sc.pl.umap(ad, color='cell_type', ax=axs[0], show=False)
sc.pl.umap(ad, color='PAX5', ax=axs[1], show=False)
sc.pl.umap(ad, color='PAX5_activity', ax=axs[2], cmap='coolwarm', vmin=-2, vmax=2)
f.tight_layout()
```
